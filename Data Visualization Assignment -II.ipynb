```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = sns.load_dataset("taxis")
```

```python
print(df.head())
```

```python
print(df.isnull().sum())
```

```python
df['distance'].fillna(df['distance'].median(), inplace=True)
```

```python
df['fare'].fillna(df['fare'].mean(), inplace=True)
```

```python
df['tip'].fillna(df['tip'].mean(), inplace=True)
```

```python
df['tolls'].fillna(df['tolls'].mean(), inplace=True)
```

```python
df['total'].fillna(df['total'].mean(), inplace=True)
```

```python
df['payment'].fillna(df['payment'].mode()[0], inplace=True)
```

```python
df['pickup_borough'].fillna(df['pickup_borough'].mode()[0], inplace=True)
```

```python
df['pickup_zone'].fillna(df['pickup_zone'].mode()[0], inplace=True)
```

```python
df.dropna(inplace=True)
```

```python
df['pickup'] = pd.to_datetime(df['pickup'])

df.sort_values('pickup', inplace=True)

plt.figure()
plt.plot(df['pickup'], df['fare'])
plt.title("Fare over Time")
plt.xlabel("Time")
plt.ylabel("Fare")
plt.show()
```

```python
borough_fare = df.groupby('pickup_borough')['fare'].sum()

borough_fare.plot(kind='bar')
plt.title("Total Fare by Borough")
plt.xlabel("Borough")
plt.ylabel("Total Fare")
plt.show()
```

```python
payment_counts = df['payment'].value_counts()

plt.figure()
plt.pie(payment_counts, labels=payment_counts.index, autopct='%1.1f%%')
plt.title("Payment Method Distribution")
plt.show()
```

```python
plt.figure()
plt.hist(df['distance'], bins=30)
plt.title("Distance Distribution")
plt.xlabel("Distance")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure()
df.boxplot(column='tip', by='pickup_borough')
plt.title("Tip Distribution by Borough")
plt.suptitle('')
plt.xlabel("Borough")
plt.ylabel("Tip")
plt.show()
```

```python
sns.countplot(x='pickup_borough', data=df)
plt.title("Trips per Borough")
plt.show()
```

```python
sns.scatterplot(x='distance', y='fare', hue='pickup_borough', data=df)
plt.title("Distance vs Fare")
plt.show()
```

```python
corr = df[['distance','fare','tip','tolls','total']].corr()

sns.heatmap(corr, annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

```python
sns.pairplot(df[['distance','fare','tip','total','pickup_zone']], hue='pickup_zone')
plt.show()
```

```python
sns.violinplot(x='payment', y='fare', data=df)
plt.title("Fare Distribution by Payment Method")
plt.show()
```
